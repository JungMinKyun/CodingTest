```python
# 백준 2668번 숫자 고르기

# DFS니까 Stack으로 구성할 수 있지 않을까 함 그래프 그림을 형성해보자

# 그래프를 그렸을 때 cycle을 여러개 찾아줘야 함
# 모든 점? 을 돌면서 cycle이 만들어지는 과정이 생기면 ans 리스트에다가 append 시키기

# vertex, edge 구현하기
# cycle을 찾으면 탐색을 멈추도록 재귀함수 구현

# cycle의 순서? 를 출력을 할건데, 작은 순 부터 정렬을 할것임, 함수로 재귀 탐색을 할 때 작은 값부터 넣으면 되지 않을까

N=int(input()) # 입력 받아서 nums에다가 넣어 놓기
nums=[]
for i in range(N):
  nums.append(int(input()))

def dfs(st, cur, visited, c): # 방문한 곳이고 시작=끝이면 cycle에 저장하는 함수
  if visited[cur]:
    if cur == st:
      c.append(st)
  else:
    visited[cur] = True
    dfs(st, nums[cur-1], visited, c)

cn = [] # 사이클들 받아놓을 cycle 개수 만들 리스트 생성
for i in range(N):
  visited = [False] * (N + 1)
  c = []
  dfs(i+1, i+1, visited, c)
  cn.extend(c)

cn = list(set(cn))
cn.sort() #정렬

print(len(cn))
for _ in cn:
  print(_)
```

```python
# 백준 16502번 그녀를 찾아서

N=int(input())
M=int(input())
ID={'A':0, 'B':1,'C':2,'D':3}
probability=[[0,0,0,0] for _ in range(4)]

for _ in range(M):
  x, y, prob = input().split()
  probability[ID[x]][ID[y]]=float(prob)

prob=[0.25]*4
result=[0]*4

def calculate(prob, result, count):
  if count==N:
    return prob
  else:
    for __ in range(4):
      val=prob[__]
      for ___ in range(4):
        x=probability[__][___]*val
        result[___]=result[___]+x
    for _ in range(4):
      prob[_]=result[_]
      result[_]=0
    calculate(prob, result, count+1)

calculate(prob, result, 0)

for _ in prob:
  print(f'{_ *100: .2f}')
```

```python
# 백준 3085번 사탕 게임

N=int(input())
board=[]
for _ in range(N):
    board.append(list(input()))
candy=0

def HOT(board):
    global candy
    for k in range(N):
        cnt=1
        for m in range(1,N):
            if board[k][m] == board[k][m-1]:
                cnt += 1
                candy = max(cnt, candy)
            else:
                cnt=1
        cnt=1
        for m in range(1,N):
            if board[m][k] == board[m-1][k]:
                cnt += 1
                candy = max(cnt, candy)
            else:
                cnt = 1

for i in range(N):
    for j in range(N):
        if j < N-1:
            board[i][j], board[i][j+1] = board[i][j+1], board[i][j]
            HOT(board)
            board[i][j], board[i][j+1] = board[i][j+1], board[i][j]

        if i < N-1:
            board[i][j], board[i+1][j] = board[i+1][j], board[i][j]
            HOT(board)
            board[i][j], board[i+1][j] = board[i+1][j], board[i][j]

print(candy)
```

```python
def dfs(x, y, score, visited):
    global max_score

    # 목표 위치에 도달한 경우, 현재까지의 점수와 최대 점수를 비교하여 갱신
    if x == 5 and y == 5:
        max_score = max(max_score, score)
        return

    # 아래로 이동하는 경우
    if x < 5 and not visited[x+1][y]:
        visited[x+1][y] = True
        dfs(x + 1, y, score + matrix[x][y], visited)
        visited[x+1][y] = False

    # 오른쪽으로 이동하는 경우
    if y < 5 and not visited[x][y+1]:
        visited[x][y+1] = True
        dfs(x, y + 1, score + matrix[x][y], visited)
        visited[x][y+1] = False


# 입력 1 처리
matrix = []
for _ in range(5):
    row = list(map(int, input().split()))
    matrix.append(row)

max_score = 0
visited = [[False] * 6 for _ in range(6)]  # 방문 여부를 저장하는 배열
visited[1][1] = True  # 시작점 방문 처리
dfs(1, 1, matrix[0][0], visited)
print(max_score)

# 입력 2 처리
coordinates = []
for _ in range(4):
    x, y = map(int, input().split())
    coordinates.append((x, y))

# 방문한 위치의 점수를 계산할 때, 입력2에 있는 좌표를 최소 1번 이상 방문하는 조건을 반영
for x, y in coordinates:
    visited = [[False] * 6 for _ in range(6)]
    visited[1][1] = True  # 시작점 방문 처리
    dfs(x, y, matrix[x-1][y-1], visited)

print(max_score)
```
