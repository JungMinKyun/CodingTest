```python
# 인건비 - 고정, 제품 생산값 - 제품 * 개수
# 제품가격이 잇을 것임
# 생산대수를 늘려갈 때, 수입이 비용보다 많아지는 순간 (손익분기점) 찾는 프로그램 찾기

# 인건비, 개당 생산비용, 개당 판매가 입력
a, b, c=map(int, input().split())
prod_income=c-b
# 개당 순이익이 0이 되면 손익분기점이 안나오니 -1 출력
if prod_income==0:
  print(-1)
else:
# '생산 사전비용 / 개당 순이익' 의 몫에 1을 더하기 (Break even Point, 손익분기점)
  BEP=a//prod_income+1
  if BEP >=0:
    print(BEP)
  else:
    print(-1)

# for문을 사용해 작성하면 너무 오래걸리고, 3 2 1 이 입력되었을 때 무한히 돌게 됨, 그냥 수학 계산으로 출력하자
```

```python
# 분수 행렬을 지그재그 순서로 찾기

# 세로 관점에서 1 2 1 / 1 2 3 4 3 2 1 / 1 2 3 4 5 4 3 2 1 ... 이렇게 진행
# 가로 관점에서 1 / 1 2 3 2 1 / 1 2 3 4 5 4 3 2 1 / 1 2 3 4 5 6 7 6 5 4 3 2 1 ... 이렇게 진행

N=int(input())
a=[]
for num in range(1, N+1):
  num=2*num
  for j in range(1,num+1):
    a.append(j)
    if len(a) > N*5:
      break
  for k in range(num-1, 0, -1):
    a.append(k)
    if len(a) > N*5:
      break
b=[]
for num in range(1, N+1):
  num=2*num-1
  for j in range(1,num+1):
    b.append(j)
    if len(b) > N*5:
      break
  for k in range(num-1, 0, -1):
    b.append(k)
    if len(b) > N*5:
      break
print('{0}/{1}'.format(b[N-1],a[N-1]))

## 논리에는 문제가 없으나, 리스트를 만들어 저장하는 과정에서 시간을 초과해버림
## 리스트를 저장하지 않고 찾는 풀이가 필요
```

```python
# 분수 행렬을 지그재그 순서로 찾기
# 가로 세로를 나누지 않고
# [1/1], [1/2, 2/1], [3/1, 2/2, 1/3], [1/4, 2/3, 3/2, 4/1]로 놓게 되면
# 즉, 홀수번째 bracket 안에 분자는 감소, 분모는 증가하는 추세를 보임 (짝수 bracket은 증감 추세가 거꾸로)
X=int(input())
line=1
while X>line:
  X-=line
  line+=1
if line%2==0:
  a=X
  b=line-X+1
else:
  a=line-X+1
  b=X
print(a,'/',b,sep='')
```

```python
# 달팽이 오르기 문제, a는 낮에 올라가는 길이 b는 잘때 내려가는 길이, v는 올라갈 총 길이
# v-a 의 값을 하루에 올라가는 길이로 나눈담에 몫에 따라 1 or 2를 더해준다

a, b, v = map(int,input().split())
real=v-a
if real/(a-b) > real//(a-b):
  print((real//(a-b))+2)
elif real/(a-b) == real//(a-b):
  print((real//(a-b))+1)
```

```python
# k층, m호에 살면, k-1층의 1~m호 사람을 합한만큼 살아야함
# 파스칼 삼각형을 돌려놓은 상태와 비슷함 (조합문제?)
# 조합을 사용한다면 (k+m) Comb (K+1)으로 계산하면 됨

N=int(input())
for i in range(N):
  head=1
  body1=1
  body2=1
  k=int(input())
  n=int(input())
  for j in range (1,k+n+1):
    head=head*j
  for l in range (1, k+2):
    body1=body1*l
  for m in range (1,n):
    body2=body2*m
  print(int(head/(body1*body2)))
```

```python
# 문제에서 주어진 코드
def recursion(s, l, r, count):
    if l >= r:
      return 1, count+1
    elif s[l] != s[r]:
      return 0, count+1
    else: return recursion(s, l+1, r-1, count+1)

def isPalindrome(s):
    return recursion(s, 0, len(s)-1, 0)

# 주어진 코드를 뜯어보자
# isPalindrome 함수에서, s에 string, 처음 인덱스(0), 마지막 인덱스(길이-1), 이후 recursion 함수로 진입
# recursion 에서는 재귀를 통해, 인덱스를 양 끝에서 좁혀 나갈 것이다,
# 이때 elif에서 좁혀오는 인덱스의 값이 다른 순간이 오면 0을 출력
# 계속 좁히다가, 인덱스 값이 같아지거나 역전되는 순간이 오면 Palindrome Number 라는 뜻이니 1을 출력한다.
# 즉, 여기서 recursion이 일어난 수는 양끝에서 부터 글자 비교를 했을 때, 판별이 되는 순간까지의 횟수가 되는 것이다.

# 함수에 횟수를 세줄 매개변수를 추가하는 식으로 접근해보자.
# count 매개변수를 추가해, recursions 에 돌 때 1씩 추가되도록 하였더니 성공

N=int(input())
for _ in range(N):
  string=input()
  print(*isPalindrome(string))
```

```python
N,S,M=map(int, input().split())
vols=list(map(int,input().split()))
vl=[S-vols[0],S+vols[0]]
for i in range(len(vols)):
  for j in range(len(vl)):
    lx = vl[j]-vols[i]
    hx = vl[j]+vols[i]
    if hx>0 or hx<=M:
      vl.append[hx]
    if lx>0 or lx<=M:
      vl.append[lx]
```

```python
# vols라는 볼륨 범위 리스트
# vl이라는 i번째에 대한 볼륨 값들 리스트 (즉 , 최신화된 S에 대한 리스트)
# vl을 계속해서 최신화 할건데, - vols의 i번째 값을 S에 쁠마해준 값을 새롭게 넣어주는 것,

# 일단 vl이라는 S가 담긴 리스트를 만들어 놓고,
# vols의 i번째 값을 vl 내부 값에 전부 쁠마를 해서 조건 만족하는 애들로 새롭게 vl을 만들고 싶은데,

# 그래서 지금 생각하는게, S값들이 담긴 리스트와, vols의 i번째 값을 넣으면 그거에 대한 새로운 S값들의 리스트를 만들어주는 함수를 만들어보려고 함

# 리스트와 V값을 넣으면 리스트 값에 전부 V를 쁠마해서 0보다 크고 M이하인 애들만 리스트에 담는함수를 만들었음
def newest(Slist, V):
  vlist=[]
  for i in Slist:
    lx=i-V
    hx=i+V
    if hx>=0 and hx<=M:
      vlist.append(hx)
    if lx>=0 and lx<=M:
      vlist.append(lx)

  Slist=list(set(vlist))
  return Slist

#입력 두줄
N,S,M=map(int, input().split())
vols=list(map(int,input().split()))

# 초기값 넣는 것과, 리스트가 empty, 즉 -1를 출력하게 되는 경우를 고려해줘야함

# 현재 음량 리스트에서 V[i] 값의 쁠마를 계산해서 새로운 리스트를 만들고자 함 (덮어쓰기)
preL=[S]

for i in vols:
  preL=newest(preL, i)

if len(preL)==0:
  print(-1)
else:
  print(max(preL))
```

```python
# 백준 19638 센티와 마법 뿅망치 문제

import heapq

N, H, T = map(int, input().split())
height=[]
for i in range(N):
  PH=int(input())
  height.append(PH)
trial=0
max_h = [-x for x in height]
heapq.heapify(max_h)

while T > 0:
    if -max_h[0] < H:
        break
    tt = -heapq.heappop(max_h)
    if tt == 1:
      new_height=1
    else:
      new_height = tt // 2
    heapq.heappush(max_h, -new_height)
    T -= 1
    trial += 1

if -max_h[0] < H:
    print("YES")
    print(trial)
else:
    print("NO")
    print(-max_h[0])
```

```python
# 백준 2580번 스도쿠

board=[]
z_loc=[]
for _ in range(9):
    row=list(map(int,input().split()))
    for __ in range(9):
        if row[__]==0:
            z_loc.append([_,__])
    board.append(row)

def check(x,y):
    for i in range(9):
        if i != x and board[i][y] == board[x][y]:
            return False
        if i != y and board[x][i] == board[x][y]:
            return False
    for i in range(x//3*3, (x//3+1)*3):
        for j in range(y//3*3, (y//3+1)*3):
            if (i != x or j != y) and board[i][j] == board[x][y]:
                return False
    return True

def sdoku(N):
    if N == len(z_loc):
        return True
    x, y = z_loc[N]
    for k in range(1, 10):
        board[x][y] = k
        if check(x, y) and sdoku(N+1):
            return True
    board[x][y] = 0
    return False

sdoku(0)

for row in board:
    print(*row)
```

```python
# 백준 14888번 연산자 끼워넣기

N=int(input())
nums=list(map(int,input().split()))
tools=list(map(int,input().split()))
ord=[]

def cal(a,b,x):
  if x==0:
    return a+b
  elif x==1:
    return a-b
  elif x==2:
    return a*b
  else:
    if a*b<0:
      return -(abs(a)//abs(b))
    else:
      return a//b

for i,k in zip(tools,[0,1,2,3]):
  for _ in range(i):
    ord.append(k)

from itertools import permutations

ans=list(permutations(ord,sum(tools)))

# print(ans, len(ans),sep='\n')

res=nums[0]
for i in range(len(ans[0])):
  res=cal(res,nums[i+1],ans[0][i])
  mim=res
  mam=res

for elm in ans:
  res=nums[0]
  for i in range(len(elm)):
    res=cal(res,nums[i+1],elm[i])
  mim=min(mim, res)
  mam=max(mam, res)

print(mam, mim, sep='\n')
```

```python
# 백준 14889번 스타트와 링크

N = int(input())
board = [list(map(int,input().split())) for _ in range(N)]
visited = [False for _ in range(N)]
INF = 2147000000
res = INF

def DFS(L,idx):
    global res
    if L == N//2:
        A = 0
        B = 0
        for i in range(N):
            for j in range(N):
                if visited[i] and visited[j]:
                    A += board[i][j]
                elif not visited[i] and not visited[j]:
                    B +=board[i][j]
        res = min(res, abs(A-B))
        return
    for i in range(idx,N):
        if not visited[i]:
            visited[i] = True
            DFS(L+1,i+1)
            visited[i] = False

DFS(0,0)
print(res)
```
